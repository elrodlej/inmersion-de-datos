```python
local_edificio_lote = (inmuebles.Tipo == 'Local')| (inmuebles.Tipo =='Edificio')|(inmuebles.Tipo =='Lote')
tres_inmuebles = inmuebles[local_edificio_lote]
tres_inmuebles.head(10)
```

```python
font = {'family': 'serif',
        'color':  'purple',
        'weight': 'bold',
        'size': 14}
```

```python
#histograma utilizando solo 3 tipos de inmuebles/refinar el grafico, letras, colores, labels / sacar una conclusion sobre la informacion
plt.figure(figsize=(10,8)) #Determina el tamaño de la figura
grafica = sns.histplot(data=tres_inmuebles, x='Precio_Millon', kde=True, hue='Tipo') #kde kernet distribution, para la linea de distribucion/ para ver tipo de valor se usa hue
titulo = grafica.set_title('Distribucion de valores de locales, lotes y edificios de Bogota')
grafica.set_xlabel('Precio en millones', fontdict=font)
grafica.set_ylabel('Cantidad de inmuebles', fontdict=font)
plt.xlim((90,16000)) #rango de valores obtenidos del eje x
plt.ylim((0,12)) #limitar valores en el eje y

plt.show()
```
